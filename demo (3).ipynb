```python
from azureml.core import Workspace, Dataset

subscription_id = '353b9ff5-223a-4dc4-853e-825bc329d30c'
resource_group = 'shakalaka'
workspace_name = 'pikachu'

workspace = Workspace(subscription_id, resource_group, workspace_name)

dataset = Dataset.get_by_name(workspace, name='bca')
dataset.to_pandas_dataframe()
```

```python
data = dataset.to_pandas_dataframe()
```

```python
crs = data.corr() 
print(crs)
```

```python
# Experiments -> contains RUNS
from azureml.core import Experiment
exp = Experiment(workspace=workspace, name='cancer_exp')
# Hyperparameters, Algorithms, train-test split, randomization -> belong to RUN!
```

```python
diagnosis = data.loc[:,'diagnosis'] #y
# x-> input 
inputs = data.loc[:,['id','radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean'] ]
```

```python
encoding_logic = lambda value: 1 if value=='M' else 0
y = diagnosis.map(encoding_logic)
y[:20]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(inputs,diagnosis, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.linear_model import SGDClassifier # gradient descent
from sklearn.svm import SVC, LinearSVC # graph
from sklearn.naive_bayes import GaussianNB # probability
from sklearn.tree import DecisionTreeClassifier # TREE based 
from sklearn.ensemble import RandomForestClassifier 
# I need some METRICS from which i could compare these 3 algorithms and select which was better or worse
import time 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
algos = {'KNN':KNeighborsClassifier(),'RFPlain': RandomForestClassifier(),'RF5':RandomForestClassifier(n_estimators=5),'RF200':RandomForestClassifier(n_estimators=200), 'Dtree':DecisionTreeClassifier(),'SGD':SGDClassifier(), 'LinearSVC':LinearSVC(), 'SVC':SVC(), 'Sherlock':GaussianNB()}
time_error = 0
import sklearn.externals import joblib
# LOOP as EXPERIMENTS -> EACh sub-loop is an individual run
for name,algo in algos.items():
    run = exp.start_logging()
    #print('Starting algo: ' + name)
    run.log("algo_name", name)
    starttime = time.time()
    model = algo
    model.fit(xtrain, ytrain)
    predictions = model.predict(xtest)
    endtime = time.time()
    accur = accuracy_score(predictions, ytest)
    cv_score = cross_val_score(model,inputs, diagnosis, cv=5 ) 
    timetaken = endtime- starttime - time_error
    #print('Time taken by '+ name +' is: ' + str(timetaken))
    run.log("time", float(timetaken))
    run.log("acc", float(accur))
    # save the model in every run 
    model_name =  name + '.pkl'
    path_name = 'outputs/' + model_name
    joblib.dump(value=model,filename=path_name)
    run.upload_file(name=model_name, path_or_stream=path_name)
    run.complete()
```
